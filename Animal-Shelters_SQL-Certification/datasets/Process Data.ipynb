```python
# Import pandas and read in raw data
import pandas as pd
animals = pd.read_csv('animals.csv')
size = pd.read_csv('size_costs.csv')
spons = pd.read_csv('sponsored_animals.csv')
location = pd.read_csv('location_costs.csv')
```

```python
## Reduce size of animals DataFrame, drop a couple of columns, and reset the index
animals_new = animals.sample(frac=0.1, random_state=42).\
    drop(columns=['outcome', 'outcomesubtype']).\
    reset_index(drop=True)
```

```python
# Grab a small sample of the animals DataFrame as the sponsored animals
sponsor_new = animals_new.sample(frac=0.18, random_state=42).drop(columns=['birthdate', 'color'])
```

```python
# Write two new processed files to csv
animals_new.to_csv('animal_data.csv', index=False)
sponsor_new.to_csv('sponsored_pets.csv', index=False)
```

```python
# Create sizeid column
animaltype = animals_new['animaltype']
weight = animals_new['weight']
sizeid = []
for i in range(len(animaltype)):
 if (animaltype[i] == 'Dog') & (weight[i]>=31.0):
    sizeid.append('DL')
 elif (animaltype[i] == 'Dog') & (30.0>=weight[i]>=11.0):
    sizeid.append('DM')
 elif (animaltype[i] == 'Dog') & (weight[i]<=10.0):
    sizeid.append('DS')
 elif (animaltype[i] == 'Cat') & (weight[i]>=8.0):
    sizeid.append('CL')   
 elif (animaltype[i] == 'Cat') & (7.0>=weight[i]>=6.0):
    sizeid.append('CM')  
 elif (animaltype[i] == 'Cat') & (weight[i]<=5.0):
    sizeid.append('CS') 
 elif (animaltype[i] == 'Bird') & (weight[i]>=1.1):
    sizeid.append('BL') 
 elif (animaltype[i] == 'Bird') & (1.1>weight[i]>=0.7):
    sizeid.append('BM')
 else:
    sizeid.append('BS')
```
